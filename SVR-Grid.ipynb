```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset.columns
```

```python
indep=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
depn=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV    # use GridSearchCrossValidation lib
from sklearn.svm import SVR  # lib for SVM & SVR                         
param_grid={'kernel':['linear','poly','rbf','sigmoid','precomputed'],
            'C':[10,100,1000,2000,3000],
            'gamma':['auto','scale']}  #plable parametter for hyper tunning
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)  #import(lib),True for find good model,
grid.fit(indep,depn)  #fitting the model for grid
```

```python
re=grid.cv_results_
print("The R_score value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
age_input=int(input("Age:"))
Bmi_input=float(input("BMI"))
Childran_input=int(input("NO childran"))
sex_male_input=int(input("sex Male 0 or 1:"))
smoker_yes_input=int(input("smoker Yes 0 or 1:"))
```

```python
Future_prediction=grid.predict([[age_input,Bmi_input,Childran_input,sex_male_input,smoker_yes_input]])
print("Future_prediction",Future_prediction)
```
